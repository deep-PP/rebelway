```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("netflix1.csv")
```

```python
df.loc[df["title"]=="Squid Game"]
```

```python
# to change Squid Game country to South Korea, but if Squid Game country is wrong we can assume that other countries will be likewise wrong

# df.loc[df["title"]=="Squid Game","country"]="South Korea"
# df.loc[df["title"]=="Squid Game"]
```

```python
series_df = df
series_df.head()
```

```python
series_df.info()
```

```python
series_df.drop(["show_id","date_added","director"], inplace=True, axis=1)
```

```python
series_df["type"].unique()
```

```python
#drop movies from dataframe

series_df = series_df[series_df["type"].str.contains("Movie") == False]
```

```python
series_df.head()
```

```python
series_df.drop(["type"], inplace=True, axis=1)
```

```python
#replacing "Not Given" values as nan to drop them in a easier way

series_df = series_df.replace({"Not Given": np.nan})
```

```python
series_df.info()
```

```python
series_df.dropna(inplace=True)
```

```python
series_df.info()
```

one hot encoding

```python
one_hot = pd.get_dummies(series_df["country"])
series_df = series_df.join(one_hot)
series_df.drop(["country"], inplace=True, axis=1)
```

```python
series_df["rating"].unique()
```

```python
one_hot = pd.get_dummies(series_df["rating"])
series_df = series_df.join(one_hot)
series_df.drop(["rating"], inplace=True, axis=1)
```

```python
series_df["duration"].unique()
```

```python
series_df["duration"]= series_df["duration"].str.extract('(\d+)').astype(int)
series_df["duration"].unique()
```

```python
series_df["listed_in"] =series_df["listed_in"].str.split(',').apply(lambda x: list(map(lambda val: val.replace(" ", ""), x)))
```

```python
series_df.reset_index(drop=True, inplace=True)
```

```python
# listed_in contains multiple genres for each tv series, MultiLabelBinarizer allows to create one hot encoding of each genre avoiding duplicate

from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

genre = pd.DataFrame(mlb.fit_transform(series_df["listed_in"]), columns=mlb.classes_)
genre
```

```python
series_df = series_df.join(genre)
```

```python
series_df.drop(["listed_in"], inplace=True, axis=1)
```

one hot encoding end

```python
series_df.head()
```

```python
series_df.set_index("title", inplace=True)
series_df.head()
```

similarity and corr

```python
series_df = series_df[series_df.notna()]
series_df.head()
```

```python
series_df = series_df.astype('int')
```

```python
squid_game =  series_df.loc['Squid Game']
squid_game
```

```python
squid_game_similarity = series_df.corrwith(squid_game, axis=1, method="kendall")
```

```python
series_corr = pd.DataFrame(squid_game_similarity, columns=['Correlation'])
series_corr
```

```python
series_corr.dropna(inplace=True)
```

```python
series_corr.sort_values('Correlation', ascending=False).head(9)
```

```python
reccomandation = series_corr.sort_values('Correlation', ascending=False).head(9)

reccomandation = df[df["title"].isin(reccomandation.index)]
reccomandation
```
