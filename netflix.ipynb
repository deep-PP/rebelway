```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
series_df = pd.read_csv("netflix1.csv")
```

```python
series_df.head()
```

```python
series_df.info()
```

```python
series_df.drop(["show_id","date_added","director"], inplace=True, axis=1)
```

```python
series_df["type"].unique()
```

```python
series_df = series_df[series_df["type"].str.contains("Movie") == False]
```

```python
series_df.head()
```

```python
series_df.drop(["type"], inplace=True, axis=1)
```

```python
series_df = series_df.replace({"Not Given": np.nan})
```

```python
series_df.info()
```

```python
series_df.dropna(inplace=True)
```

```python
series_df.info()
```

```python
def set_col_mapping(df, col_type):
    object_columns = df.select_dtypes(include=[col_type]).columns.tolist()
    print(object_columns)
    for col in object_columns:
        res = {}
        for x, value in enumerate(df[col].value_counts().index.tolist()):
            res[value] = x
        print(res)
        df[col] = df[col].map(res)

    return df
```

```python
tmp = set_col_mapping(series_df.drop(["title"], axis=1), 'object')
series_df = pd.concat([series_df["title"], tmp], axis=1)
```

```python
series_df.head()
```

```python
series_df["title"].str.contains("nan").value_counts()
```

```python
series_df = series_df[series_df["title"].str.contains("nan") == False]
```

```python
series_df.set_index("title", inplace=True)
series_df.head()
```

similarity and corr

```python
series_df[["Squid Game" in s for s in series_df.index]]
```

```python
squid_game =  series_df[["Squid Game" in s for s in series_df.index]]
```

```python
squid_game_similarity =series_df.corrwith(squid_game, axis=1)
```

```python
series_corr = pd.DataFrame(squid_game_similarity, columns=['Correlation'])
series_corr
```

```python
series_corr.dropna(inplace=True)
```

```python
series_corr.sort_values('Correlation', ascending=False).head(10)
```
